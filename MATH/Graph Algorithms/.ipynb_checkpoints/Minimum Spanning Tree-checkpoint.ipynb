## Minimum Spanning Trees
A minimum spanning tree is a subset of the edges $M \subseteq E$ of a connected, edge-weighted undirected graph $G = (V, E)$ that connects all the vertices together, without any cycles and with the minimum possible total edge weight. That is, it is a spanning tree whose sum of edge weights is as small as possible.

```python
import networkx as nx

G = nx.Graph()
G.add_weighted_edges_from([('A', 'B', 3.0), ('A', 'D', 4.0), ('A', 'E', 8.0)])
G.add_weighted_edges_from([('B', 'C', 2.0), ('B', 'E', 8.0)])
G.add_weighted_edges_from([('C', 'D', 2.0), ('C', 'E', 8.0)])

pos=nx.circular_layout(G)
edge_weights = {edge : G.get_edge_data(*edge)["weight"] for edge in G.edges()}
min_tree = nx.minimum_spanning_tree(G, algorithm='kruskal')
colored_min_tree_edges=['red' if edge in min_tree.edges else 'black' for edge in G.edges]

nx.draw(G, pos, node_color='green', edge_color=colored_min_tree_edges, with_labels=True, width=3, node_size=2000)
edge_labels = nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_weights,
    font_color='green',
    font_size=15
)
```
