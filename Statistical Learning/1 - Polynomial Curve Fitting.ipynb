Suppose we are interested in predicting the functional relationship between some some real-valued random vector $\mathbf{X_i} \in \mathbb{R}^n$ (commonly called **features**, **predictors**, **variables**, **input vector**) and each cooresponding real-valued output $\mathbf{Y} \in \mathbb{R}^m$ (commonly called the **quantitative response**, **signal**, **output vector**) on the basis of $N$ observations of $\mathbf{X_i} \in \{\mathbf{X_1, X_2, \ldots, X_p}\}$ known as the **training set**. 


## The Philosophy of Supervised Regression

As described above, in supervised regression, we assume the existence of some functional relationship $f : \mathbb{R}^n \to \mathbb{R}^m$ that sends a set of predictors $X = \{\mathbf{X_1, X_2, \ldots, X_p}\} \subsetneq \mathbb{R}^n$ to its image $Y = \{\mathbf{Y_1, Y_2, \ldots, Y_p}\} \subsetneq \mathbb{R}^m$.

$$
\begin{equation}
Y_i = f(X_i) + \epsilon_i
\label{eq:systemic} \tag{1}
\end{equation}
$$

In $\eqref{eq:systemic}$, $f$ is the **systematic information that $X$ provides about $Y-$** the relationship between the *known* variables of $X$ and the output $Y$. Since $Y$ may actually be determined by more variables than in $X$, the error terms $\epsilon_i \sim N(0),$ is the **information about $Y$ that could not possibly be inferred in $f(X)$** on the basis of the fact that no training set is likely to ever be *omniscient*. Note, in an ideal model, $\epsilon$ should follow the same distribution as the Gaussian function.

If we knew the actual value of $f$, we would not need supervised regression. The goal therefore is to come up with a consistent way to cook up some relation $\hat f$ on the basis of training data which reliably approximates the true relationship described by $f$. In addition to cooking up some approximation, the technique should also cook up a rigorous way to measure "*how closely*" $\hat f$ predicts $f$ and "*how closely*" $\hat f$ explains the systematic information that $X$ provides about $Y$ (from a topolgical point of view, "closely" actually refers to a notion of physical proximity).

## Polynomial Fitting

To motivate a number of key concepts and methods for approximating the systematic relationship between $X$ and $Y$ and to motivate techniques for evaluating the "closeness" of our approximation to the true relationship, we shall consider an artificial example where the relationship between inputs and outputs is completely known (again, this would not be possible in practice). 

Let $\mathbb{P(R)} \ni f : X \to Y$, defined by following equation, be the **systematic information that $X \in \mathbb{R}$ provides about $Y \in \mathbb{R}$**,

$$
\begin{equation}
f(x) = \sin(2 \pi x)
\label{eq:relationship} \tag{2}
\end{equation}
$$

for all $x \in X(\Omega)$. 

In Matplotlib, the relation is visualized as follows,

```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.sin(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(15, 6), dpi=80)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
plt.show()
```
